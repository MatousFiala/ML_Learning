```python
import pandas as pd
import numpy as np
import sklearn.linear_model
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml("mnist_784", version = 1, as_frame=False)
```

```python
X, y = mnist["data"], mnist["target"].astype(np.uint8)


# Train-test split
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:] 
```

```python
# Eights classifier
y_is_eights_train, y_is_eights_test = (y_train == 8), (y_test == 8)


SGD_eights_classifier = sklearn.linear_model.SGDClassifier(random_state=90)

SGD_eights_classifier.fit(X_train, y_is_eights_train)
```

```python
from sklearn.model_selection import cross_val_predict
import sklearn.metrics

y_train_predictions = cross_val_predict(SGD_eights_classifier, X_train, y_is_eights_train, cv=3)

sklearn.metrics.confusion_matrix(y_is_eights_train, y_train_predictions)
```

```python
import sklearn.metrics
print(sklearn.metrics.precision_score(y_is_eights_train,y_train_predictions),

sklearn.metrics.recall_score(y_is_eights_train,y_train_predictions))
```
